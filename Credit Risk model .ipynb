# A dataset is typically considered imbalanced if one class has significantly more samples than the other.

```python
from collections import Counter 
```

```python
counter_value = Counter(data['default.payment.next.month'])
majority_class = max(counter_value.values())
minority_class = min(counter_value.values())
```

```python
imbalance_ratio  = majority_class/minority_class
```

```python
data['default.payment.next.month'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='coolwarm')
plt.ylabel("")
plt.title('Imbalance Ratio')
plt.show()
```

We found that its imbalance class, there are few techniques to handle this

1. Resampling: Oversampling the minority class (SMOTE) or undersampling the majority class.
2. Using Weighted Models: Adjusting class weights in algorithms like RandomForestClassifier or LogisticRegression.
3. Using Anomaly Detection: If one class is rare, treat it as an anomaly detection problem.

Method -->	Pros-->	Cons-->	Best for

SMOTE (Oversampling)--->Keeps all majority data, generates realistic samples	May create synthetic noise	Small datasets, where removing majority data is risky

Random Oversampling	--> Simple, fast, easy to implement	Can lead to overfitting	Small datasets

Random Undersampling --> Reduces training time, balances dataset	Can lose useful data	Large datasets

Cluster Undersampling --> Keeps important majority samples	More computationally expensive	Large datasets, when you want to retain key majority patterns
